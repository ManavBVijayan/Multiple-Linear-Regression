```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data
np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
y = 4 + 3 * X1 + 2 * X2 + np.random.randn(100, 1)
```

```python
# Combine independent variables
X = np.c_[X1, X2]

# Add a bias term for the intercept
X_b = np.c_[np.ones((100, 1)), X]
```

```python
# Use the normal equation to calculate the coefficients
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Print the coefficients
print("Intercept and Coefficients:", theta_best.ravel())
```

```python
# Make predictions
X_new = np.array([[1, 2, 3]])  # Example input for prediction
X_new_b = np.c_[np.ones((1, 1)), X_new[:, 1:]]  # Add bias term
y_pred = X_new_b.dot(theta_best)

print("Predicted Output:", y_pred[0, 0])
```

```python
# Plot the data and regression plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')

# Create a meshgrid for the regression plane
x1_plane, x2_plane = np.meshgrid(np.linspace(0, 2, 100), np.linspace(0, 3, 100))
y_plane = theta_best[0] + theta_best[1] * x1_plane + theta_best[2] * x2_plane

# Plot the regression plane
ax.plot_surface(x1_plane, x2_plane, y_plane, alpha=0.5, cmap='viridis')

plt.show()
```
